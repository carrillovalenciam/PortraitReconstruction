# Apply Gaussian Filter to Portrait Dataset
## Getting images


```python
import numpy as np
from skimage.filters import gaussian
import os
import matplotlib.pyplot as plt
```

```python
input_folder = './Train/'
output_folder = './Train_Gaussian/'
```

```python
sample_1 = plt.imread(input_folder + os.listdir(input_folder)[0])
print(sample_1.shape)
plt.imshow(sample_1)
```

```python
#We try to apply the gaussian filter to this single image using skimage
#For a first try we will apply sigma=3 to filter
gaussian_3_sample_1 = gaussian(sample_1, sigma=3,multichannel=True)
gaussian_5_sample_1 = gaussian(sample_1, sigma=5,multichannel=True)
gaussian_10_sample_1 = gaussian(sample_1, sigma=10, multichannel = True)
```

```python
plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.imshow(gaussian_3_sample_1)
plt.subplot(1,3,2)
plt.imshow(gaussian_5_sample_1)
plt.subplot(1,3,3)
plt.imshow(gaussian_10_sample_1)
```

From the above images, we can see that the blur is notable from sigma=3. With this we will restrict
the gaussian filter from sigma 1 to 10.


```python
from PIL import Image
import PIL
```

```python
#I have the original images in the 'Train' folder. I will save them into the 'Train_Gaussian' folder
#that will be used later as input for the network
for file_name in os.listdir(input_folder):
    original_image = plt.imread(input_folder + file_name)  
    
    random_sigma = np.random.randint(1,11)
    #input_image = gaussian_filter(real_image, sigma = random_sigma)
    blur_image = gaussian(original_image, sigma = random_sigma, multichannel=True)
    image_to_save = Image.fromarray((blur_image*255).astype(np.uint8))
    image_to_save.save(output_folder + file_name)
```
